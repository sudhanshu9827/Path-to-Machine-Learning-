# Five-Number Summary

The **Five-Number Summary** is a set of five descriptive statistics that provides a concise overview of a dataset's distribution, center, and spread. It is commonly used in exploratory data analysis and is visually represented using a **box plot**.

The five values, arranged from smallest to largest, are:

1. **Minimum**  
   The smallest value in the dataset.

2. **First Quartile (Q1)**  
   The 25th percentile, meaning 25% of the data lies below this value.

3. **Median (Q2)**  
   The middle value (50th percentile) that divides the dataset into two equal halves.

4. **Third Quartile (Q3)**  
   The 75th percentile, meaning 75% of the data lies below this value and 25% lies above it.

5. **Maximum**  
   The largest value in the dataset.

The Five-Number Summary helps in understanding the distribution, identifying potential outliers, and analyzing the overall spread of the data.

```python
sns.boxplot(x=df['Age'])
```
