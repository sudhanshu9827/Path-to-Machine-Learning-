```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 30, 35, np.nan],
    'Salary': [50000, np.nan, 70000, 90000],
    'Experience': [2, 5, 8, 10]
})
```

```python
df
```

```python
imputer = IterativeImputer(random_state=42)
```

```python
df_imputed = imputer.fit_transform(df)
```

```python
df_imputed = pd.DataFrame(
    df_imputed,
    columns=df.columns
)

print(df_imputed)
```
